# Hero and Publisher

```python
dfsh[["Hero", "Publisher"]] = dfsh["Hero|Publisher"].str.split('|',expand=True)
dfsh.head()
```

```python
# drop original 'Hero|Publisher' column
dfsh = dfsh.drop(columns = ["Hero|Publisher"])

# check
dfsh.head(2)
```

```python
# create an order that makes more sense
dfsh = dfsh[["Publisher","Hero","Gender","Race","Alignment","Hair color","Eye color","Skin color", "Measurements"]]
dfsh.head()
```

# Heights and Weights

```python
# check type of 'Measurements' column
measurement = dfsh.loc[0, 'Measurements']
print(type(measurement))
print(measurement)
```

```python
# convert strings in dictionary to using double instead of
# single quotes for json.loads compatibility
dfsh['Measurements'] = dfsh['Measurements'].str.replace("'", '"')

# check
measurement = dfsh.loc[0, 'Measurements']
print(measurement)
```

```python
# use .apply with json.loads to convert strings to dicts
dfsh['Measurements'] = dfsh['Measurements'].apply(json.loads)

# check
measurement = dfsh.loc[0, 'Measurements']
print(type(measurement))
```

```python
# unpack 'Measurements' column (dictionaries) into separate
# columns for each dictionary item
height_weight = dfsh['Measurements'].apply(pd.Series)

# check
height_weight
```

```python
# concatenate new columns to original df
dfsh = pd.concat((dfsh, height_weight), axis = 1)

# check
dfsh.head(2)
```

```python
# drop original column
dfsh = dfsh.drop(columns = ['Measurements'])

# check
dfsh.head(2)
```

```python
# change 'Height' and 'Weight' column names
dfsh.rename(columns = {'Height': 'Height (cm)',
                      'Weight': 'Weight (kg)'}, 
            inplace = True)

# check
dfsh.head(2)
```

```python
# take off 'cm' and 'kg' from cell values in 'Height' and 'Weight'
to_replace = [' cm', ' kg']

for char in to_replace:
    dfsh['Height (cm)'] = dfsh['Height (cm)'].str.replace(char, 
                                                          '', 
                                                          regex = False)
    dfsh['Weight (kg)'] = dfsh['Weight (kg)'].str.replace(char, 
                                                          '', 
                                                          regex = False)
    
# check
dfsh[['Height (cm)', 'Weight (kg)']].head(2)
```

```python
# convert 'Height (cm)' and 'Weight (kg)' columns from object-
# types to numeric

dfsh['Height (cm)'] = dfsh['Height (cm)'].astype(float)
dfsh['Weight (kg)'] = dfsh['Weight (kg)'].astype(float)

# check
dfsh.info()
```

# Powers

```python
# check type of powers['Powers'] column
power = dfsp.loc[0, 'Powers']
print(type(power))
print(power)
```

```python
# split string on comma
dfsp['powers_split'] = dfsp['Powers'].str.split(',')

# check
dfsp[['Powers', 'powers_split']].head(2)
```

```python
exploded = dfsp.explode('powers_split')
exploded[['hero_names', 'Powers', 'powers_split']].head(5)
```

```python
# save unique values from exploded column
cols_to_make = exploded['powers_split'].dropna().unique()

# check
cols_to_make
```

```python
# create new column for each unique value in list
# populate cells with True or False

frames = [dfsp]

for col in cols_to_make:
    frame = pd.DataFrame()
    frame[col] = dfsp['Powers'].str.contains(col)
    frames.append(frame)
    merged_powers = pd.concat(frames, axis = 1)
    
# check
merged_powers.head()
```

```python
# drop 'Powers' and 'powers_split' columns
merged_powers.drop(columns = ['Powers', 'powers_split'], inplace = True)

# check
merged_powers.head()
```

```python
# change 'hero_names' column name to match info['Hero']
merged_powers.rename(columns = {'hero_names': 'Hero'}, 
                     inplace = True)

# check
merged_powers.head(2)
```

```python
# check info for merged_powers before merging
merged_powers.info()
```

```python
# check info for info before merging
dfsh.info()
```

```python
# combine two dataframes on info['Hero'] and merged_powers['Hero']
hero_df = pd.merge(dfsh, 
                   merged_powers, 
                   on = 'Hero')

# check
hero_df.head()
```

```python
# check
hero_df.info()
```

# II. Use your combined DataFrame to answer the following questions.

Compare the average weight of super powers who have Super Speed to those who do not.
What is the average height of heroes for each publisher?

```python
hero_df.groupby('Super Speed')['Weight (kg)'].mean().round(2)
```

From the information gathered by the dataset, the Super Hero who has Super Speed weighs an average of 129.4 KG.
The Super Hero who does not have Super Speed weighs an average of 101.77 KG.
An interesting further break down would be to compare the genders of the Super Heroes, then their ages.

```python
hero_df.groupby('Publisher')['Height (cm)'].mean().round(2)
```

The rest of the publishers all into this ascending order: Shueisha, Dark Horse, Unknown, Team Epic TV, Star Trek, DC Comics, and Marvel Comics.
